```python
#Importer les librairies necessaires
import numpy as np
import pandas as pd
import csv
import os
import seaborn as sb
```

```python
dataset=pd.read_csv("Original_Dataset.csv",index_col=0)
```

```python
dataset.shape
```

```python
for i in range (1 , 31) :
    if (i not in [1,5,10,15,20,25,30]):
        del dataset["LC_X_"+str(i)]
        del dataset["LC_Y_"+str(i)]
        del dataset["LD_"+str(i)]
        del dataset["RC_X_"+str(i)]
        del dataset["RC_Y_"+str(i)]
        del dataset["RD_"+str(i)]
```

```python
dataset.shape
```

```python
for i in range (1,6) :
    for j in range (1 , 31) :
        del dataset["LF_X_"+str(i)+"_"+str(j)]
        del dataset["LF_Y_"+str(i+1)+"_"+str(j)]
        del dataset["RF_X_"+str(i)+"_"+str(j)]
        del dataset["RF_Y_"+str(i)+"_"+str(j)]
```
