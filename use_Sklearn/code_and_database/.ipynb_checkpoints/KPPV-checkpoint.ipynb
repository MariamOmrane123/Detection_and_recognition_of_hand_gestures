```python
#Importer les librairies necessaires
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

%matplotlib inline
```

```python
#charger votre base de données
dataset=pd.read_csv("Hand_Gesture_Recognition_Dataset.csv")
```

```python
data=dataset.iloc[:,:-2].values
target=dataset.iloc[:,-2].values
```

```python
from sklearn.model_selection import train_test_split
data_train,data_test,target_train,target_test=train_test_split(data,target,test_size=0.2,random_state=0)
```

```python
print("data_train shape: {}".format(data_train.shape))
print("target_train shape: {}".format(target_train.shape))
```

```python
print("data_test shape: {}".format(data_test.shape))
print("target_test shape: {}".format(target_test.shape))
```

```python
from sklearn import neighbors
knn = neighbors.KNeighborsClassifier(n_neighbors=3)
knn.fit(data_train, target_train)
```

```python
target_pred=knn.predict(data_test)
```

```python
df=pd.DataFrame({"actual":target_test,"predicted":target_pred})
df
```

```python
print("test set accuracy: {:2f}".format(knn.score(data_test,target_test)))
error=1-knn.score(data_test,target_test)
print("erreur:%f" %error)
```

```python
from sklearn import metrics
print("Mean Squared Error:",metrics.mean_squared_error(target_test,target_pred))
print("Mean Absolute Error:",metrics.mean_absolute_error(target_test,target_pred))
```

```python
errors=[]
for k in range(2,15):
    knn=neighbors.KNeighborsClassifier(k)
    errors.append(100*(1-knn.fit(data_train,target_train).score(data_test,target_test)))
plt.plot(range(2,15),errors,'o-')
plt.show()
```
