```python
#Importer les librairies necessaires
import numpy as np
import pandas as pd
import csv
import os
import seaborn as sb
```

```python
dataset=pd.read_csv("Origin_Hand_Gesture_Recognition_Dataset.csv")
```

```python
dataset=dataset.iloc[:300,:]
```

```python
dataset.shape
```

```python
dataset
```

```python
#add a Gesture_Class column 
#0:ZoomIn  1:ZoomOut  2:MoveLeft  3:MoveRight  4:MoveDown  5:MoveUp  
dataset['Gesture_Class'] = 0
for i in range(50):
    dataset['Gesture_Class'][i] = 0
for i in range(50,100):
    dataset['Gesture_Class'][i] = 1
for i in range(100,150):
    dataset['Gesture_Class'][i] = 2
for i in range(150,200):
    dataset['Gesture_Class'][i] = 3
for i in range(200,250):
    dataset['Gesture_Class'][i] = 4
for i in range(250,300):
    dataset['Gesture_Class'][i] = 5
    
```

```python
dataset
```

```python
# get a list of columns
cols = list(dataset)

# move the column Gesture_Class
cols.insert(cols.index('Gesture'), cols.pop(cols.index('Gesture_Class')))

# use ix to reorder
dataset = dataset.ix[:, cols]

#rename column
dataset.rename(columns={'Gesture': 'Gesture_Name'}, inplace=True)
```

```python
dataset.to_csv('Modified_Hand_Gesture_Recognition_Dataset.csv') 
```
